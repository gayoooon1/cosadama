```
import nltk # natural language toolkit
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```
# Ïó∞ÏäµÏö© Îç∞Ïù¥ÌÑ∞ train 
train = [('i like you','pos'),
         ('i hate you', 'neg'),
         ('you like me', 'neg'),
         ('i like her',' pos')]
# ÏßëÌï© ÏûêÎ£åÌòï ÏÇ¨Ïö© Ï§ëÎ≥µ ÏóÜÏï†Í∏∞ ÏúÑÌï®
```

```
for sentence in train: 
  for word in word_tokenize(sentence[0]): 
    print(word)
```

```
all_words = set(word.lower() for sentence in train # word.lower() ÏÜåÎ¨∏ÏûêÎ°ú Î≥ÄÌôò
                for word in word_tokenize(sentence[0]))
all_words
```

```
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1])
          for x in train]
t
```

ÌõàÎ†® ÏãúÌÇ® Í≤ÉÏùÑ Î∞îÌÉïÏúºÎ°ú classifier ÏûëÎèô!

```
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```
test_sentence = 'i like MeRui'
test_sent_features = {word.lower(): 
                      (word in word_tokenize(test_sentence.lower()))
                      for word in all_words}

test_sent_features
```

test sentenceÏóê Î∂ÑÎ•òÌïú Îã®Ïñ¥Í∞Ä ÏûàÎäîÏßÄ Î®ºÏ†Ä ÌôïÏù∏Ìïú Îí§, classifier ÏûëÎèô

```
classifier.classify(test_sent_features)
```

**ÌïúÍ∏Ä**

ÌïúÍ∏ÄÎ°ú classifier ÏûëÎèô ÏãúÌÇ§Í∏∞ ÏúÑÌï¥ÏÑúÎäî ÌòïÌÉúÏÜå Î∂ÑÏÑùÏùÑ Ìï¥ÏïºÌïòÎäîÎç∞, ÏïàÌïòÎ©¥ Ïñ¥ÎñªÍ≤å Îê†Íπå?

```
from konlpy.tag import Okt
```

```
pos_tagger = Okt()
```

```
train = [('Î©îÎ¶¨Í∞Ä Ï¢ãÏïÑ', 'pos'),
         ('Í≥†ÏñëÏù¥ÎèÑ Ï¢ãÏïÑ', 'pos'),
         ('ÎÇú ÏàòÏóÖÏù¥ ÏßÄÎ£®Ìï¥', 'neg'),
         ('Î©îÎ¶¨Îäî Ïù¥ÏÅú Í≥†ÏñëÏù¥Ïïº', 'pos'),
         ('ÎÇú ÎßàÏπòÍ≥† Î©îÎ¶¨Îûë ÎÜÄÍ±∞Ïïº', 'pos')]
```

```
all_words = set(word.lower() for sentence in train
                        for word in word_tokenize(sentence[0]))
all_words
```

Ï°∞ÏÇ¨Í∞Ä Î∂ôÏñ¥ÏÑú Í∞ôÏùÄ Îã®Ïñ¥Ïù∏Îç∞ Îã§Î•∏ Îã®Ïñ¥Î°ú Ïù∏ÏãùÌïúÎã§.

```
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1])
                                                        for x in train]
t
```

```
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```
test_sentence = 'ÎÇú ÏàòÏóÖÏù¥ ÎßàÏπòÎ©¥ Î©îÎ¶¨Îûë ÎÜÄÍ±∞Ïïº'
```

```
test_sent_features = {word.lower(): 
                      (word in word_tokenize(test_sentence.lower()))
                      for word in all_words}
test_sent_features
```

```
classifier.classify(test_sent_features)
```

ÌòïÌÉúÏÜå Ï†úÍ±∞ ÏïàÌï¥ÏÑú negÎ°ú ÌåêÎ≥ÑÌïú Í≤ÉÏùÑ Î≥º Ïàò ÏûàÎã§.

**ÌòïÌÉúÏÜå Î∂ÑÏÑù ÌõÑ classify**

```
def tokenize(doc): # Ìï®ÏàòÏóê ÌÉúÍ∑∏Î•º Î∂ôÏó¨Ï£ºÎ©¥ Ï¢ãÎã§Í≥† Ìï©ÎãàÎã§...
  return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

```
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```
tokens = [t for d in train_docs for t in d[0]]
tokens
```

Ìï¥Îãπ Î¨∏Ïû•Ïóê ÏûàÎäî ÌòïÌÉúÏÜåÍ∞Ä tokens ÏïàÏóê Îì§Ïñ¥ÏûàÎäîÏßÄ ÌôïÏù∏ÌïòÎäî Ìï®Ïàò

```
def term_exists(doc): 
  return {word: (word in set(doc)) for word in tokens}
```

```
train_xy = [(term_exists(d), c) for d,c in train_docs]
train_xy
```

```
classifier = nltk.NaiveBayesClassifier.train(train_xy)
```

```
test_sentence = [('ÎÇú ÏàòÏóÖÏù¥ ÎßàÏπòÎ©¥ ÏΩîÏÇ¨Îã§ÎßàÎûë ÎÜÄÍ±∞Ïïº')]
```

```
# ÌòïÌÉúÏÜå Î∂ÑÏÑù 
test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```
test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

```
classifier.classify(test_sent_features)
```

```
# ÌÖçÏä§Ìä∏Ïùò ÌäπÏßïÏùÑ Ï∂îÏ∂úÌïòÎäî Î™®Îìà
from sklearn.feature_extraction.text import CountVectorizer
```

ÍµêÏû¨ 316Ï™Ω Ï∞∏Í≥†Ìï¥ÏÑú Î≤°ÌÑ∞Ìôî Ïù¥Ìï¥ÌïòÍ∏∞

```
import scipy as sp 

def dist_raw(v1, v2):
  delta = v1 - v2 
  return sp.linalg.norm(delta.toarray())
```

`np.linalg.norm()` Í∞Ä Î≤°ÌÑ∞Ïùò Îã®ÏúÑ normÏùÑ Íµ¨ÌòÑÌï¥ Ï£ºÎäî Ïó≠Ìï†ÏùÑ ÌïúÎã§.

Ïù¥Í≤ÉÏùÑ ÌÜ†ÎåÄÎ°ú Îã§Ïãú Í±∞Î¶¨Î•º Íµ¨ÌïòÍ∏∞
Ïù¥Î≤àÏóêÎäî Îã§Î•∏ Ìï®ÏàòÎ•º Ï†ÅÏö©Ìï¥ÏÑú Í±∞Î¶¨Î•º Íµ¨Ìï¥ Î≥∏Îã§. `dist_norm(v1, v2)`

```
def dist_norm(v1, v2): 
  v1_normalized = v1 / sp.linalg.norm(v1.toarray()) # ÎÜà Í∏∞Ï§ÄÏúºÎ°ú...Î≥ÄÌôò...?
  v2_normalized = v2 / sp.linalg.norm(v2.toarray())

  delta = v1_normalized - v2_normalized 

  return sp.linalg.norm(delta.toarray())
```

* IDF = Î™®Îì† Î¨∏ÏÑúÏóê Í∞ôÏùÄ Îã®Ïñ¥Í∞Ä ÏûàÏùÑ Í≤ΩÏö∞ Î¨∏ÏÑú Í∞Ñ ÎπÑÍµêÏóêÏÑúÎäî Ï§ëÏöîÌïú Îã®Ïñ¥Í∞Ä ÏïÑÎãàÎùºÍ≥† ÌåêÎã®, ÌïµÏã¨ Ïñ¥ÌúòÏùºÏàòÎäî ÏûàÏßÄÎßå! ÏïÑÎßà Îã§ ÏûàÏúºÎãàÍπå Î≥ÑÎ°ú Ï§ëÏöîÌïòÏßÄ ÏïäÎã§Í≥† ÏÉùÍ∞ÅÌïòÎäî Í≤É Í∞ôÎã§.

= **TF-IDF Í∏∞Î∞òÏùò Î≤°ÌÑ∞Ìôî**

[More about TF-IDF](https://charsyam.wordpress.com/2017/04/08/%EC%9A%A9%EC%96%B4-%EC%A0%95%EB%A6%AC-%EC%9E%85-%EA%B0%9C%EB%B0%9C%EC%9E%90%EB%A5%BC-%EC%9C%84%ED%95%9C-tf-idf/)
