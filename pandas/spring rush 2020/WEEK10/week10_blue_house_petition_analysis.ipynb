### 2. DataFrameÏóê Ï†ïÎ≥¥ Îã¥Í∏∞

```python
pet_title = []
pet_start = []
pet_end = []
pet_ppl = []
pet_ans = []
pet_url = []
pet_ars = []

#def get_data() :
for pgn in range(1,13):
    urlbd = 'https://www1.president.go.kr/petitions/answer?page='
    pg = urlopen(urlbd + str(pgn))
    soup = BeautifulSoup(pg, "html.parser")
    tmp_x = soup.find_all('div', class_='ans_name')

    for n in range(0,len(tmp_x)):
      tmp = tmp_x[n]
      pet_title.append(tmp.find(class_='ans_name_title').a.get_text())
      # Ï†úÎ™©Îßå Í∞ÄÏ†∏ÏôÄÏïº ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê a Ï∂îÍ∞Ä
      # ÏïàÍ∑∏Îü¨Î©¥ Ï≤≠ÏõêÏãúÏûëÏùº Îì± Î™®Îì† Îç∞Ïù¥ÌÑ∞Í∞Ä Ìè¨Ìï®ÎêòÏñ¥ ÎÇòÏò¥
      pet_start.append(tmp.find(class_='ans_list1_2').get_text()[2:-2])
      pet_end.append(tmp.find(class_='ans_list2_2').get_text()[2:-2])
      pet_ppl.append(tmp.find(class_='ans_list3_2').get_text()[2:-2])
      pet_ans.append(tmp.find(class_='ans_list4_2').get_text()[2:-2])
      pet_ars.append('https://www1.president.go.kr' + str(tmp.find('a')['href']))

data = {'Ï†úÎ™©':pet_title, 'Ï≤≠ÏõêÏãúÏûëÏùº':pet_start,
        'Ï≤≠ÏõêÏ¢ÖÎ£åÏùº':pet_end,'Ï∞∏Ïó¨Ïù∏Ïõê':pet_ppl,'Ï≤≠ÏõêÎãµÎ≥ÄÏùº':pet_ans,'URL':pet_ars}
df = pd.DataFrame(data)
df.head()
```

```python
import re 
import time
from tqdm import tqdm_notebook # PYTHON ÏßÑÌñâÌëúÏãúÎ∞î ÎßåÎì§Í∏∞
from urllib.request import urlopen
from bs4 import BeautifulSoup

category =[]
content = []
answer = []
person = []

for n in tqdm_notebook(petition.index): 
    url = urlopen(petition['URL'][n]) 
    time.sleep(1)
    soup = BeautifulSoup(url, "html.parser")
    tmp_x = soup.find('ul',class_='petitionsView_info_list').get_text()
    p0 = re.compile('\s')
    splited = p0.split(tmp_x) #tmp_y.li.get_text()
    category.append(splited[1:5][0][4:]) # Ïπ¥ÌÖåÍ≥†Î¶¨ Í∞ÄÏ†∏Ïò§Í∏∞
    person.append(splited[1:5][3][3:]) # Ï≤≠ÏõêÏù∏ Í∞ÄÏ†∏Ïò§Í∏∞
    
    tmp_y = soup.find_all('div',class_='View_write')
    if len(tmp_y) > 1:
        content_raw = tmp_y[1].get_text()
    else:
        content_raw = tmp_y[0].get_text()
    p1 = re.compile('\r|\t|\n|\xa0')
    content_splited = p1.split(content_raw)
    content.append(''.join(content_splited)) # Ï≤≠ÏõêÎÇ¥Ïö© Í∞ÄÏ†∏Ïò§Í∏∞
    
    tmp_z = soup.find_all('div', class_='pr_tk25')
    if len(tmp_z) > 1:
        ans_raw = tmp_z[1].get_text()
    else:
        ans_raw = tmp_z[0].get_text()
    ans_splited = p1.split(ans_raw)
    ans_splited = ' '.join(ans_splited)
    answer.append(ans_splited[7:]) # Ï≤≠ÏõêÎãµÎ≥Ä Í∞ÄÏ†∏Ïò§Í∏∞
```

```python
import nltk
from konlpy.tag import Okt
okt = Okt()
```

```python
humanRights = petition[petition['Ïπ¥ÌÖåÍ≥†Î¶¨'] == 'Ïù∏Í∂å/ÏÑ±ÌèâÎì±']
humanRights.shape
```

```python
humanRights_text = ''

for each_line in humanRights['Ï≤≠ÏõêÎÇ¥Ïö©']: 
  humanRights_text = humanRights_text + each_line + '\n'
```

```python
tokens_ko = okt.morphs(humanRights_text)
tokens_ko
```

```python
stop_words = ['.','ÏùÑ','Ïù¥','Ïùò','Î•º','Ïóê',',','Í∞Ä','Îì§','ÏùÄ','Îäî','ÏúºÎ°ú','Ìïú','ÎèÑ',
              'Ïàò','ÏóêÏÑú','Î°ú','Í≤É','Í∑∏','Í≥º','Ï†ú','ÏûÖÎãàÎã§','Ìï†','ÌïòÍ≥†','Ï†Å',"'",'ÌïòÎäî',
              'Ìï©ÎãàÎã§','ÏôÄ','ÏóêÍ≤å','Í≥†','Ïù∏',')','?','ÌïòÏó¨','Îì±','Ï†Ä','ÏûàÏäµÎãàÎã§','Îßê','ÍπåÏßÄ',
              '(','1','Í∑∏Î¶¨Í≥†','Îã§','2','Îßå','Îî∏','ÌñàÏäµÎãàÎã§','Ïïà','Îêú','Î™ª','Ïùº','Îçî','ÏúÑ',
              'ÏûàÎäî','Ìï¥','\n','ÎòêÌïú','Î™Ö','ÌïòÏßÄ','..','Ï†ïÎßê','Î¨º','ÎßéÏùÄ','"',
              'Ï§ë','3','Í≤å','ÎÑàÎ¨¥','A','Ïûê','Ïù¥Îü∞','Îïå','ÎêòÏßÄ','ÏúÑÌï¥','ÏóêÎäî','ÏóÜÎäî',
              'Î∞è','Ïöî','ÎïåÎ¨∏','Í¥ÄÎ†®','ÎåÄÌïú','Ï†Ñ','ÎùºÍ≥†','ÎêòÏóàÏäµÎãàÎã§','ÌïòÍ∏∞','ÎÇò','ÌõÑ','Í∞ôÏùÄ','Ìï¥ÏÑú',
              'Í∏Ä','ÎêòÏñ¥','Ï£º','Ï∞®','Ìïò','ÎÇ¥','ÏóÜÏù¥','ÎåÄÌï¥','Ïûò','ÌïòÏßÄÎßå','Í≤∞Íµ≠']

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]

ko = nltk.Text(tokens_ko, name='Ïù∏Í∂å/ÏÑ±ÌèâÎì± Ï≤≠ÏõêÎÇ¥Ïö©')
ko.vocab().most_common(50)
```

### 5. Word2VecÏùÑ Ïù¥Ïö©Ìï¥ Îã®Ïñ¥Í∞Ñ Ïú†ÏÇ¨ÏÑ± Ï∞æÍ∏∞

word2vec -> word to vector Ïù¥ÎùºÎäî ÎúªÏùºÎìØ „Öã„Öã„Öã„Öã Îã®Ïñ¥Í∞Ñ Ïú†ÏÇ¨ÏÑ± ÌôïÏù∏ÌïòÍ∏∞!

```python
import gensim 
from gensim.models import word2vec
```

```python
okt = Okt()
results = []

for line in humanRights['Ï≤≠ÏõêÎÇ¥Ïö©']: 
  # ÌòïÌÉúÏÜå Î∂ÑÏÑù 
  malist = okt.pos(line, norm=True, stem=True) #Ï™ºÍ∞úÎ≤ÑÎ†§
  #print(malist)
  r = []

  # word2vec Ïã§Ìñâ
  for word in malist: 
    if not word[1] in ['Josa', 'Eomi','punctuation']:
      r.append(word[0])

  r1 = (" ".join(r)).strip()
  results.append(r1)
  #print(r1)
```

```python
data_file = 'human_rights.data'
with open(data_file, 'w', encoding='utf-8') as fp: 
  fp.write("\n".join(results)) # Ï†Ä-Ïû•~
```

```python
data = word2vec.LineSentence(data_file)
model = word2vec.Word2Vec(data, size=200, window=10, hs=1, min_count=2, sg=1)

model.save('human_rights.model')
```
