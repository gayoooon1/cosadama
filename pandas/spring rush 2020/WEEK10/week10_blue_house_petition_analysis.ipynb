```python
pet_title = []
pet_start = []
pet_end = []
pet_ppl = []
pet_ans = []
pet_url = []
pet_ars = []

#def get_data() :
for pgn in range(1,13):
    urlbd = 'https://www1.president.go.kr/petitions/answer?page='
    pg = urlopen(urlbd + str(pgn))
    soup = BeautifulSoup(pg, "html.parser")
    tmp_x = soup.find_all('div', class_='ans_name')

    for n in range(0,len(tmp_x)):
      tmp = tmp_x[n]
      pet_title.append(tmp.find(class_='ans_name_title').a.get_text())
      # ì œëª©ë§Œ ê°€ì ¸ì™€ì•¼ í•˜ê¸° ë•Œë¬¸ì— a ì¶”ê°€
      # ì•ˆê·¸ëŸ¬ë©´ ì²­ì›ì‹œì‘ì¼ ë“± ëª¨ë“  ë°ì´í„°ê°€ í¬í•¨ë˜ì–´ ë‚˜ì˜´
      pet_start.append(tmp.find(class_='ans_list1_2').get_text()[2:-2])
      pet_end.append(tmp.find(class_='ans_list2_2').get_text()[2:-2])
      pet_ppl.append(tmp.find(class_='ans_list3_2').get_text()[2:-2])
      pet_ans.append(tmp.find(class_='ans_list4_2').get_text()[2:-2])
      pet_ars.append('https://www1.president.go.kr' + str(tmp.find('a')['href']))

data = {'ì œëª©':pet_title, 'ì²­ì›ì‹œì‘ì¼':pet_start,
        'ì²­ì›ì¢…ë£Œì¼':pet_end,'ì°¸ì—¬ì¸ì›':pet_ppl,'ì²­ì›ë‹µë³€ì¼':pet_ans,'URL':pet_ars}
df = pd.DataFrame(data)
df.head()
```

```python
import re 
import time
from tqdm import tqdm_notebook # PYTHON ì§„í–‰í‘œì‹œë°” ë§Œë“¤ê¸°
from urllib.request import urlopen
from bs4 import BeautifulSoup

category =[]
content = []
answer = []
person = []

for n in tqdm_notebook(petition.index): 
    url = urlopen(petition['URL'][n]) 
    time.sleep(1)
    soup = BeautifulSoup(url, "html.parser")
    tmp_x = soup.find('ul',class_='petitionsView_info_list').get_text()
    p0 = re.compile('\s')
    splited = p0.split(tmp_x) #tmp_y.li.get_text()
    category.append(splited[1:5][0][4:]) # ì¹´í…Œê³ ë¦¬ ê°€ì ¸ì˜¤ê¸°
    person.append(splited[1:5][3][3:]) # ì²­ì›ì¸ ê°€ì ¸ì˜¤ê¸°
    
    tmp_y = soup.find_all('div',class_='View_write')
    if len(tmp_y) > 1:
        content_raw = tmp_y[1].get_text()
    else:
        content_raw = tmp_y[0].get_text()
    p1 = re.compile('\r|\t|\n|\xa0')
    content_splited = p1.split(content_raw)
    content.append(''.join(content_splited)) # ì²­ì›ë‚´ìš© ê°€ì ¸ì˜¤ê¸°
    
    tmp_z = soup.find_all('div', class_='pr_tk25')
    if len(tmp_z) > 1:
        ans_raw = tmp_z[1].get_text()
    else:
        ans_raw = tmp_z[0].get_text()
    ans_splited = p1.split(ans_raw)
    ans_splited = ' '.join(ans_splited)
    answer.append(ans_splited[7:]) # ì²­ì›ë‹µë³€ ê°€ì ¸ì˜¤ê¸°
```

```python
import nltk
from konlpy.tag import Okt
okt = Okt()
```

```python
humanRights = petition[petition['ì¹´í…Œê³ ë¦¬'] == 'ì¸ê¶Œ/ì„±í‰ë“±']
humanRights.shape
```

```python
humanRights_text = ''

for each_line in humanRights['ì²­ì›ë‚´ìš©']: 
  humanRights_text = humanRights_text + each_line + '\n'
```

```python
tokens_ko = okt.morphs(humanRights_text)
tokens_ko
```

```python
stop_words = ['.','ì„','ì´','ì˜','ë¥¼','ì—',',','ê°€','ë“¤','ì€','ëŠ”','ìœ¼ë¡œ','í•œ','ë„',
              'ìˆ˜','ì—ì„œ','ë¡œ','ê²ƒ','ê·¸','ê³¼','ì œ','ì…ë‹ˆë‹¤','í• ','í•˜ê³ ','ì ',"'",'í•˜ëŠ”',
              'í•©ë‹ˆë‹¤','ì™€','ì—ê²Œ','ê³ ','ì¸',')','?','í•˜ì—¬','ë“±','ì €','ìˆìŠµë‹ˆë‹¤','ë§','ê¹Œì§€',
              '(','1','ê·¸ë¦¬ê³ ','ë‹¤','2','ë§Œ','ë”¸','í–ˆìŠµë‹ˆë‹¤','ì•ˆ','ëœ','ëª»','ì¼','ë”','ìœ„',
              'ìˆëŠ”','í•´','\n','ë˜í•œ','ëª…','í•˜ì§€','..','ì •ë§','ë¬¼','ë§ì€','"',
              'ì¤‘','3','ê²Œ','ë„ˆë¬´','A','ì','ì´ëŸ°','ë•Œ','ë˜ì§€','ìœ„í•´','ì—ëŠ”','ì—†ëŠ”',
              'ë°','ìš”','ë•Œë¬¸','ê´€ë ¨','ëŒ€í•œ','ì „','ë¼ê³ ','ë˜ì—ˆìŠµë‹ˆë‹¤','í•˜ê¸°','ë‚˜','í›„','ê°™ì€','í•´ì„œ',
              'ê¸€','ë˜ì–´','ì£¼','ì°¨','í•˜','ë‚´','ì—†ì´','ëŒ€í•´','ì˜','í•˜ì§€ë§Œ','ê²°êµ­']

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]

ko = nltk.Text(tokens_ko, name='ì¸ê¶Œ/ì„±í‰ë“± ì²­ì›ë‚´ìš©')
ko.vocab().most_common(50)
```

### 5. Word2Vecì„ ì´ìš©í•´ ë‹¨ì–´ê°„ ìœ ì‚¬ì„± ì°¾ê¸°

word2vec -> word to vector ì´ë¼ëŠ” ëœ»ì¼ë“¯ ã…‹ã…‹ã…‹ã…‹ ë‹¨ì–´ê°„ ìœ ì‚¬ì„± í™•ì¸í•˜ê¸°!

```python
import gensim 
from gensim.models import word2vec
```

```python
okt = Okt()
results = []

for line in humanRights['ì²­ì›ë‚´ìš©']: 
  # í˜•íƒœì†Œ ë¶„ì„ 
  malist = okt.pos(line, norm=True, stem=True) #ìª¼ê°œë²„ë ¤
  #print(malist)
  r = []

  # word2vec ì‹¤í–‰
  for word in malist: 
    if not word[1] in ['Josa', 'Eomi','punctuation']:
      r.append(word[0])

  r1 = (" ".join(r)).strip()
  results.append(r1)
  #print(r1)
```

```python
data_file = 'human_rights.data'
with open(data_file, 'w', encoding='utf-8') as fp: 
  fp.write("\n".join(results)) # ì €-ì¥~
```

```python
data = word2vec.LineSentence(data_file)
model = word2vec.Word2Vec(data, size=200, window=10, hs=1, min_count=2, sg=1)

model.save('human_rights.model')
```
