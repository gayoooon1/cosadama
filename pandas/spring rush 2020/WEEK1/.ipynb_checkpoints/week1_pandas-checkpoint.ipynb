```python
#데이터프레임 삭제 'drop' 열, 행 모두 삭제 가능
#DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
#axis=1 칼럼 열은 축이 있다고 생각하기

ttnc_df.drop(1)
#ttnc_df.drop(0) #인덱스 삭제
ttnc_df.drop('SibSp',axis=1) #칼럼 삭제
#원본데이터 적용 X, 적용하려면 inplace=True

# 다수의 인덱스 삭제하기 
ttnc_df.drop([0,1,2])
```

```python
#슬라이싱, 인덱싱 가능
# row로 슬라이싱
#ttnc_df[2:5]

# 불린 인덱싱(조건을 넣어 인덱싱하기): Survived 컬럼 값이 0인 row만 가져와라!
ttnc_df[ttnc_df['Survived']==0].head(3)

# loc[]을 통해서도 가능
ttnc_df.loc[ttnc_df['Age'] > 60, ['Name','Age']].head()
##칼럼 잘라서 가져올 때 불린 인덱싱은 대괄호 2번, loc,iloc은 한번 ['Name','Age'] 참고##
```

```python
##예제##
#타이타닉 호에서 사망자, 1등급 실, 30세 미만인 사람들의 명단#
ttnc_df[(ttnc_df['Survived']==0) & (ttnc_df['Pclass'] == 1) & (ttnc_df['Age']<30)]
```

```python
#DataFrame 정렬 하기
# by 안에 특정 칼럼이름을 넣으면 그 칼럼 기준으로 정렬, sort_values() 이용!!
ttnc_sorted = ttnc_df.sort_values(by=['Age'])
ttnc_sorted
```

```python
#Aggregation 함수 적용
#min(), max(), sum(), count() 등과 같은 aggregation 함수
#모든 칼럼에 aggregation을 적용

ttnc_df['Age'].mean()
```

```python
#groupby() 적용하기
#칼럼 기준 다른 칼럼으로 묶기 group을 by sth~으로 묶는다고 생각

ttnc_df.groupby('Survived')[['Age','Name']].count()

#결손 데이터 처리하기
#isna()은 NaN 여부를 확인, fillna()는 NaN 값을 다른 값으로 대체
#drop 때랑 마찬가지로 inplace=True 적용해야 실제 데이터 수정됨
ttnc_df['Age'].fillna('0')
ttnc_df['Cabin'].fillna('C000')
```

```python
#apply lambda =함수를 한줄로~
plus_10=lambda x:x+10
print(plus_10(3))
a=[10,20,30]
plus_10_upgrade=map(lambda x:x+10,a)
list(plus_10_upgrade)
##그냥 plus_10_upgrade만 치면 ERROR <map at 0x24320105948>, print 함수 사용불가##

ttnc_df['Elderly'] = ttnc_df['Age'].apply(lambda x: 'Elderly' if x > 65 else 'Not Elderly')
ttnc_df[['Age','Elderly']].head(8)
#조건 단 람다
```

```python
#데이터 프레임 병합
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])

####<인덱스 기준으로 합치기>###
#concat(): default가 axis=0이므로 열방향으로 데이터를 합침
df1_2_3=pd.concat([df1,df2,df3])


# key를 지정해 여러개의 인덱스를 만들기 ㄱㄴ
result = pd.concat([df1,df2,df3], keys=['x','y','z'])
result

df1_3=pd.concat([df1,df3],axis=1)
df1_3
#공통 인덱스에 맞추거나, 한 df 기준으로 맞출 수도 있음, 새로 만들 수도 있음

###<칼럼 기준으로 합치기>###
#merge():합병이란 뜻으로 알고~
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
a=pd.merge(left, right)
a
#교집합, 합집합, 어떠한 칼럼 기준으로 합, 모두 가능!
```
