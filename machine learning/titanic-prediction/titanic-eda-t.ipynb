# Titanic EDA

### Data dictionary
* __survival__ 0 = No, 1= Yes
* __pclass__ A proxy for socio-economic status 1 = 1st, 2 = 2nd, 3=3rd
* __sex__ male, female
* __SibSp__ The dataset defines family relations
    * Sibling = brother, sister, stepbrother, stepsister
    * Spouse = husband, wife (mistresses and fiancés were ignored)
* __parch__ The dataset defines family relations
    * Parent = mother, father
    * Child = daughter, son, stepdaughter, stepson
    * Some children travelled only with a nanny, therefore parch=0 for them
* __ticket__ ticket number
* __fare__ 티켓값
* __cabin__ 선실
* __embarked__ Port of Embarkation/ C = Cherbourg, Q = Queenstown, S = Southampton

## 1. Preprocessing

### 1-1. 준비

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('6주차 데이터 파일/train.csv')
df.shape
```

```python
df.head()
```

```python
#계산을 위해 성별을 0과 1로 바꿔준다
sex_mapping = {'male':0, 'female':1}
df['Sex'] = df['Sex'].map(sex_mapping)
df.head()
```

```python
#이름에서 Mr, Mrs, Miss 추출
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df.head()
```

```python
df['Title'].value_counts()
```

```python
#Mr, Mrs, Miss만 남기고 나머지 이름들은 3으로 처리한다
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
df['Title'] = df['Title'].map(title_mapping)
    
    
df.head()
```

```python
#객실 넘버에서 알파벳만 남긴다
df['Cabin'] = df['Cabin'].str[:1]
df.head()
```

```python
#가족 수를 합쳐서 새로운 column Family를 만들어준다
df['Family'] = df['SibSp'] + df['Parch']
df.head(20)
```

```python
#계산을 위해 object값을 int로 변형한다
df['Survived'] = df['Survived'].astype(int)
df['Pclass'] = df['Pclass'].astype(int)
df['Sex'] = df['Sex'].astype(int)
df['SibSp'] = df['SibSp'].astype(int)
df['Parch'] = df['Parch'].astype(int)
```

**5-1-4-1.중위값으로 나눠보는 경우**

```python
df["Fare"].median()
```

```python
fare_li = []
for n in df.Fare:
    if n > df.Fare.median():
        n = 1
        fare_li.append(n)
    else:
        n = 0
        fare_li.append(n)
        
df["F_level"] = fare_li
df.F_level
```

* **F_level = 1**

: 요금이 중위값 초과
    
* **F_level = 0**

: 요금이 중위값 이하


```python
F_count = df["Survived"].groupby(df["F_level"]).count()
Fare_Survived =  pd.DataFrame(df["Survived"].groupby(df["F_level"]).sum())
Fare_Survived['F_count'] = F_count
Fare_Survived["Rate"] = Fare_Survived.Survived/Fare_Survived.F_count
Fare_Survived
```

```python
sns.barplot(data = df, x = "F_level",y = Fare_Survived["Rate"])
```

* 요금이 상위 50%가 넘는 경우의 생존율이 월등히 높다
