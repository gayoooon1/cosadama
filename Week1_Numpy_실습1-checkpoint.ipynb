```python
import numpy as np
step=np.random.randint(-1,2)
walk=0
draws=[walk]
for i in range(1000):
    walk=+step
    draws.append(walk)
print(draws)
```

```python
# numpy를 모르고 파이썬 프로그래밍만으로 만들 때 
import random
position = 0
walk = [position]
for i in range(1000):
    step = 1 if random.randint(0,1) else -1
    position + step
    walk.append(position)
print(walk)
```
