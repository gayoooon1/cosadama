<h1>시군구별 성폭력 건수</h1>

```python
성폭력_raw_data = pd.read_csv(r'C:\Users\안지은\Downloads\광역시_구별_성폭력발생.csv', thousands = ',')
성폭력_raw_data
```

```python
for n in range(4,9):
    성폭력_raw_data['201'+str(n)+'년'].astype(float)
성폭력_raw_data = 성폭력_raw_data.drop(129,0)
성폭력_raw_data
```

```python
p = []
for n in range(129):
    if 성폭력_raw_data['항목.1'][n] != 성폭력_raw_data['항목.2'][n]:
        a = 성폭력_raw_data['항목.1'][n] + 성폭력_raw_data['항목.2'][n]
        p.append(a)
    elif 성폭력_raw_data['항목.1'][n] == '계':
        p.append(성폭력_raw_data['항목.1'][n])
    
    else:
        p.append(성폭력_raw_data['항목.1'][n])
성폭력_raw_data['지역'] = p        
성폭력_raw_data
```

```python
#광역시 - 시군구 합치기
for n in range(129):
    if len(성폭력_raw_data['지역'][n]) == 8:
        성폭력_raw_data['지역'][n] = 성폭력_raw_data['지역'][n][:2]+ ' '+성폭력_raw_data['지역'][n][6:]+'구'
    elif len(성폭력_raw_data['지역'][n]) == 9:
        성폭력_raw_data['지역'][n] = 성폭력_raw_data['지역'][n][:2]+ ' '+성폭력_raw_data['지역'][n][:2]+'구'
    elif len(성폭력_raw_data['지역'][n]) == 7:
        성폭력_raw_data['지역'][n] = 성폭력_raw_data['지역'][n][3:]+ ' '+ 성폭력_raw_data['지역'][n]
    else:
        성폭력_raw_data['지역'][n] = 성폭력_raw_data['지역'][n]

성폭력 = 성폭력_raw_data.drop('항목.1',1)
성폭력 = 성폭력.drop('항목.2',1)
성폭력 = 성폭력[['지역','2014년','2015년','2016년','2017년','2018년']]
성폭력
```

```python
#서울에 특별시 붙이기
for n in range(129):
    if bool('서울' in 성폭력['지역'][n]) == True:
        성폭력['지역'][n] = 성폭력['지역'][n][:2]+'특별시'+' '+성폭력['지역'][n][-3:]
    else:
        성폭력['지역'][n] = 성폭력['지역'][n]
성폭력
```

```python
#광역시 붙이기
for n in range(129):
    if len(성폭력['지역'][n]) == 6:
        성폭력['지역'][n] = 성폭력['지역'][n][:2]+'광역시'+' '+성폭력['지역'][n][-3:]
    elif len(성폭력['지역'][n]) == 7:
        성폭력['지역'][n] = 성폭력['지역'][n][:2]+'광역시'+' '+성폭력['지역'][n][-4:]
    else:
        성폭력['지역'][n] = 성폭력['지역'][n]
성폭력
```

```python
#동서남북중구 수정하기
for n in range(129):
    if bool(성폭력['지역'][n][-3:] in ['중구구', '남구구', '북구구', '서구구', '동구구'])==True:
        성폭력['지역'][n] = 성폭력['지역'][n][:-1]
    else:
        성폭력['지역'][n] = 성폭력['지역'][n]
성폭력    
```

```python
#구 -> 군
for n in range(129):
    if bool(성폭력['지역'][n][-3:] in ['달성구','강화구', '울주구', '옹진구','기장구'])==True:
        성폭력['지역'][n] = 성폭력['지역'][n][:-1] + '군'
    else:
        성폭력['지역'][n] = 성폭력['지역'][n]
성폭력    
```
